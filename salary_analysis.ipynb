# 📊 College Major Salary Analysis

This project analyzes salary trends across different college majors to understand earning potential, career growth, and risk associated with different fields.

## 📁 Dataset Overview

The dataset contains salary information for different college majors, including:

- Starting Median Salary  
- Mid-Career Median Salary  
- 90th Percentile Salary  
- Major Group (STEM, Business, HASS)  

The goal is to analyze career outcomes based on these metrics.

```python
# importing csv file

import pandas as pd
df = pd.read_csv("salaries_by_college_major.csv")
```

```python
# Majors with Highest Potential

highest_potential = clean_data.sort_values(by="Mid-Career 90th Percentile Salary", ascending=False)
highest_potential[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

## ⚙️ Feature Engineering

Created a new column **Spread**:

Spread = Mid-Career 90th Percentile Salary − Starting Salary  

This helps measure **risk vs reward** for each major.

```python
#Highest Spread of Salaries

clean_data["Spread"] = clean_data["Mid-Career 90th Percentile Salary"] - clean_data["Starting Median Salary"]
highest_spread = clean_data.sort_values(by="Spread", ascending=False)
highest_spread[["Undergraduate Major", "Spread"]].head()
```

```python
# Highest Mid-Career Salary Majors

highest_mid_career_salaries = clean_data.sort_values(by="Mid-Career Median Salary", ascending=False)
highest_mid_career_salaries[["Undergraduate Major", "Mid-Career Median Salary"]].head()
```

```python
# Getting low risk major degrees

low_risk = clean_data.sort_values('Spread')
low_risk[["Undergraduate Major", "Spread"]].head()
```

```python
# Grouping by 'Group' column

clean_data.groupby('Group').count()
```

```python
# Getting Mean

pd.options.display.float_format = '{:,.2f}'.format
clean_data.groupby('Group').mean(numeric_only=True)
```

## 📊 Data Visualization

We visualize salary trends to better understand:

- Top paying majors  
- Salary growth potential  
- Differences across major groups  

```python
import matplotlib.pyplot as plt

# Top 10 highest paying majors (mid-career)
top10 = clean_data.sort_values(by="Mid-Career Median Salary", ascending=False).head(10)

plt.figure()
plt.barh(top10["Undergraduate Major"], top10["Mid-Career Median Salary"])
plt.xlabel("Salary")
plt.ylabel("Major")
plt.title("Top 10 Highest Paying Majors (Mid-Career)")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Salary Spread (risk vs reward)
top_spread = clean_data.sort_values(by="Spread", ascending=False).head(10)

plt.figure()
plt.barh(top_spread["Undergraduate Major"], top_spread["Spread"])
plt.xlabel("Salary Spread")
plt.title("Top 10 Majors with Highest Salary Spread")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Group-wise average salary
group_avg = clean_data.groupby("Group")["Mid-Career Median Salary"].mean()

plt.figure()
group_avg.plot(kind='bar')
plt.ylabel("Average Salary")
plt.title("Average Mid-Career Salary by Major Group")
plt.show()
```

## 🔍 Key Insights

- Engineering majors dominate the highest-paying careers  
- Some majors (like Economics & Finance) show high salary growth but also higher risk  
- STEM fields have the highest average mid-career salaries  
- HASS majors tend to have lower salary growth compared to STEM  

These insights can help students choose majors based on earning potential and career stability.
